```python
import torch
import matplotlib.pyplot as plt
```

```python
def H(x):
    return -(x * torch.log(x + 1e-12)).sum()

def H_a(x, a):
    return (1/(a-1)) * (1 - (x**a).sum())
```

```python
x_list = [torch.tensor([p, 1-p]) for p in torch.linspace(0.01, 0.99, 100)]
H_list = [H(x) for x in x_list]

a_values = [1.1, 2, 3, 5, 7]
H_a_list = [[H_a(x, a) for x in x_list] for a in a_values]

plt.plot([x[0].item() for x in x_list], H_list, label='H(p)')
for i, a in enumerate(a_values):
    plt.plot([x[0].item() for x in x_list], H_a_list[i], label=f'H_(p, {a})')

plt.xlabel('p')
plt.ylabel('H')
plt.legend()
plt.show()
```
